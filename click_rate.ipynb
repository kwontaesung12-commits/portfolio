## 피처 엔지니어링

### Subtask:
기존 특성을 조합하거나 변환하여 새로운 특성을 생성하여 모델 성능 향상을 시도합니다.


**Reasoning**:
Generate new features based on the 'seq' column and add them to both training and test dataframes.



```python
# Feature Engineering based on 'seq'
def sequence_features(seq_str):
    if isinstance(seq_str, str):
        seq_list = seq_str.split(',')
        return len(seq_list), len(set(seq_list))
    else:
        return 0, 0 # Handle potential non-string values

df_train[['seq_length', 'unique_seq_items']] = df_train['seq'].apply(sequence_features).tolist()
df_test[['seq_length', 'unique_seq_items']] = df_test['seq'].apply(sequence_features).tolist()

display(df_train[['seq', 'seq_length', 'unique_seq_items']].head())
display(df_test[['seq', 'seq_length', 'unique_seq_items']].head())
```

**Reasoning**:
Create time-related features from 'hour' and 'day_of_week' and add them to both training and test dataframes.



```python
# Feature Engineering based on 'hour' and 'day_of_week'
df_train['is_weekend'] = df_train['day_of_week'].apply(lambda x: 1 if x in [6, 7] else 0)
df_test['is_weekend'] = df_test['day_of_week'].apply(lambda x: 1 if x in [6, 7] else 0)

# Convert 'hour' to numeric, handling potential errors and then to int
df_train['hour'] = pd.to_numeric(df_train['hour'], errors='coerce').fillna(-1).astype(int)
df_test['hour'] = pd.to_numeric(df_test['hour'], errors='coerce').fillna(-1).astype(int)

# Create time of day categories
def time_of_day(hour):
    if hour == -1:
        return 'unknown'
    elif 5 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    elif 18 <= hour < 22:
        return 'evening'
    else:
        return 'night'

df_train['time_of_day'] = df_train['hour'].apply(time_of_day)
df_test['time_of_day'] = df_test['hour'].apply(time_of_day)

display(df_train[['day_of_week', 'hour', 'is_weekend', 'time_of_day']].head())
display(df_test[['day_of_week', 'hour', 'is_weekend', 'time_of_day']].head())
```

**Reasoning**:
Re-prepare the features and target for model training and testing by dropping unnecessary columns and handling categorical features, including the newly created ones. Align the columns of training and test data and display the shapes to confirm.



```python
# Prepare data with new features
X = df_train.drop(['clicked', 'seq'], axis=1)
y = df_train['clicked']
X_test = df_test.drop(['ID', 'seq'], axis=1)

# Convert categorical columns to 'category' dtype, including new ones
categorical_cols = ['gender', 'age_group', 'inventory_id', 'day_of_week', 'hour', 'time_of_day']
for col in categorical_cols:
    X[col] = X[col].astype('category')
    X_test[col] = X_test[col].astype('category')

# One-hot encode categorical features
X = pd.get_dummies(X, columns=categorical_cols)
X_test = pd.get_dummies(X_test, columns=categorical_cols)

# Align columns between training and test data - crucial for consistent feature sets
train_cols = X.columns
X_test = X_test.reindex(columns=train_cols, fill_value=0)

print("Shape of X:", X.shape)
print("Shape of X_test:", X_test.shape)
```

**Reasoning**:
Apply SMOTE oversampling to the training data (X and y) to handle the class imbalance and print the shapes of the resampled data.



```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_resampled_fe, y_resampled_fe = smote.fit_resample(X, y)

print("Shape of original training data (X):", X.shape)
print("Shape of original training labels (y):", y.shape)
print("Shape of resampled training data (X_resampled_fe):", X_resampled_fe.shape)
print("Shape of resampled training labels (y_resampled_fe):", y_resampled_fe.shape)
```

**Reasoning**:
The error indicates that there are NaN values in the feature set X after feature engineering and one-hot encoding. SMOTE cannot handle NaN values. I need to check for and handle these missing values before applying SMOTE. A simple approach is to fill NaN values with the mean of the respective columns.



```python
import numpy as np

# Check for NaN values in X
print("NaN values before handling:")
print(X.isnull().sum().sum())

# Fill NaN values with the mean of the column
X = X.fillna(X.mean())

# Check for NaN values after handling
print("NaN values after handling:")
print(X.isnull().sum().sum())

# Now apply SMOTE
smote = SMOTE()
X_resampled_fe, y_resampled_fe = smote.fit_resample(X, y)

print("Shape of original training data (X):", X.shape)
print("Shape of original training labels (y):", y.shape)
print("Shape of resampled training data (X_resampled_fe):", X_resampled_fe.shape)
print("Shape of resampled training labels (y_resampled_fe):", y_resampled_fe.shape)
```

**Reasoning**:
Train a LightGBM model using the resampled training data with the new features and then predict probabilities on the test data. Check for and handle NaN values in the test data before prediction.



```python
import lightgbm as lgb

# Train the LightGBM model on the resampled data with new features
lgbm_model_fe = lgb.LGBMClassifier()
lgbm_model_fe.fit(X_resampled_fe, y_resampled_fe)

# Check for NaN values in X_test and fill them with the mean of the corresponding columns in X
print("NaN values in X_test before handling:")
print(X_test.isnull().sum().sum())

# Fill NaN values in X_test with the mean from the training data (X)
X_test = X_test.fillna(X.mean())

print("NaN values in X_test after handling:")
print(X_test.isnull().sum().sum())

# Predict probabilities using the trained model on the prepared test data
y_pred_proba_fe = lgbm_model_fe.predict_proba(X_test)[:, 1]

# Display the first few predicted probabilities
display(y_pred_proba_fe[:10])
```

**Reasoning**:
Evaluate the performance of the retrained model with the new features and resampled data using appropriate metrics for imbalanced datasets.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Convert predicted probabilities to binary predictions using a threshold (e.g., 0.5)
y_pred_fe = (y_pred_proba_fe > 0.5).astype(int)

# Calculate evaluation metrics
accuracy_fe = accuracy_score(y_test, y_pred_fe)
precision_fe = precision_score(y_test, y_pred_fe)
recall_fe = recall_score(y_test, y_pred_fe)
f1_fe = f1_score(y_test, y_pred_fe)
roc_auc_fe = roc_auc_score(y_test, y_pred_proba_fe)

# Print the evaluation metrics
print(f"Accuracy (FE + Resampled Model): {accuracy_fe:.4f}")
print(f"Precision (FE + Resampled Model): {precision_fe:.4f}")
print(f"Recall (FE + Resampled Model): {recall_fe:.4f}")
print(f"F1-Score (FE + Resampled Model): {f1_fe:.4f}")
print(f"ROC AUC Score (FE + Resampled Model): {roc_auc_fe:.4f}")
```

**Reasoning**:
The model performance is still poor, indicating the 0.5 threshold might not be optimal for this imbalanced dataset. Calculate evaluation metrics using a different threshold, such as the mean of the predicted probabilities, which is more suitable for imbalanced datasets.



```python
# Calculate evaluation metrics using the mean of predicted probabilities as the threshold
mean_val = y_pred_proba_fe.mean()
y_pred_fe_threshold = (y_pred_proba_fe > mean_val).astype(int)

accuracy_fe_threshold = accuracy_score(y_test, y_pred_fe_threshold)
precision_fe_threshold = precision_score(y_test, y_pred_fe_threshold)
recall_fe_threshold = recall_score(y_test, y_pred_fe_threshold)
f1_fe_threshold = f1_score(y_test, y_pred_fe_threshold)

print(f"Evaluation with threshold = mean of probabilities ({mean_val:.4f}):")
print(f"Accuracy (FE + Resampled Model, Mean Threshold): {accuracy_fe_threshold:.4f}")
print(f"Precision (FE + Resampled Model, Mean Threshold): {precision_fe_threshold:.4f}")
print(f"Recall (FE + Resampled Model, Mean Threshold): {recall_fe_threshold:.4f}")
print(f"F1-Score (FE + Resampled Model, Mean Threshold): {f1_fe_threshold:.4f}")
print(f"ROC AUC Score (FE + Resampled Model): {roc_auc_fe:.4f}") # ROC AUC doesn't depend on threshold
```
